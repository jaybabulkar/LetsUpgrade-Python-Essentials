# Question 1:-

## You all are Pilots, you want to land a plane safely, so altitude required for landing a plane is 1000ft., if it is less than tell the pilot to land the plane, or if it is more than but less than 5000ft. ask the pilot to "come down to 1000ft.", else if it is more than 5000ft. ask the pilot to "go around and try later". 



## Example : Input - 1000
###                         Output - Safe to land.



## Example : Input - 4500
###                         Output - Bring down to 1000.



## Example : Input - 6500
###                         Output - Turn around.


```
altitude = input ("Enter the altitude:-")                          # Here we ask the user to enter the altitude
 
altitude = int(altitude)
 
if altitude <= 1000:                                               #Now we check if the altitude is less than 1000ft.
       print ("Pilot you can land the plane.")
elif 1000 < altitude <= 5000:                                      #Now we check if the altitude is more than 1000ft. but less than 5000ft.
       print ("Pilot bring the plane down to 1000ft.")
else:                                                              #In other condition, other than those specified above.
 
       print ("Pilot go around and try later.")
```

# Question 2:-

### Using for loop please print all the prime numbers beween 1-200 using FOR LOOP and RANGE function.


### Output:-  2,3,5,7,....

```
for num in range(1,201):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num,end=" ")        
```

# OR

```
# Python programs to print all the prime numbers within the interval 1-200.

lower = 0
upper = 200

print ("Prime numbers between",lower, "and", upper, "are:")

for num in range (lower, upper + 1):
  #all the prime numbers are greater than 1.

  if num > 1:
    for i in range (2, num):
      if (num % i) == 0:
        break
      else:   
        print (num)
```
